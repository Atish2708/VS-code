```python
from dotenv import load_dotenv
load_dotenv()
```

```python
import os
api_key=os.getenv("API_KEY")
```

```python
import requests
```

```python
def get_symbol(company):
    url = "https://alpha-vantage.p.rapidapi.com/query"

    querystring = {"datatype":"json","keywords":company,"function":"SYMBOL_SEARCH"}

    headers = {
        "x-rapidapi-key": api_key,
        "x-rapidapi-host": "alpha-vantage.p.rapidapi.com"
    }

    response = requests.get(url, headers=headers, params=querystring)

    return response.json()
```

```python
get_symbol("Axis Bank")
```

```python
def get_stock_data(symbol):
    url = "https://alpha-vantage.p.rapidapi.com/query"

    querystring = {"function":"TIME_SERIES_DAILY","symbol":symbol,"outputsize":"compact","datatype":"json"}

    headers = {
    "x-rapidapi-key": api_key,
    "x-rapidapi-host": "alpha-vantage.p.rapidapi.com"
    }

    response = requests.get(url, headers=headers, params=querystring)

    return(response.json())
```

```python
get_stock_data("AXISBANK.BSE")
```

```python
data=get_stock_data("AXISBANK.BSE")
data
```

```python
import pandas as pd
df = pd.DataFrame(data['Time Series (Daily)']).T
df
```

```python
df.info
```

```python
df.index = pd.to_datetime(df.index)
df.index
```

```python
for i in df.columns:
    df[i] = df[i].astype(float)
```

```python
df
```

```python
df.to_csv("Axis Bank")
```

## Plotting the candlestick chart

```python
import plotly.graph_objects as go
```

```python
fig = go.Figure(data=[go.Candlestick(
    x = df.index,
    open = df['1. open'],
    high = df['2. high'],
    low = df['3. low'],
    close = df['4. close']
)])
fig.update_layout(width=800, height=600)
fig.show()
```
